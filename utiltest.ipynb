```python
def cvtCellCoord2ImgCoord(in_tensor, S):
    device = in_tensor.device
    i = torch.arange(S, device=device).view(1, S, 1, 1)
    j = torch.arange(S, device=device).view(1, 1, S, 1)

    x_center = (j + in_tensor[..., 0:1]) / S
    y_center = (i + in_tensor[..., 1:2]) / S

    w = in_tensor[..., 2:3]
    h = in_tensor[..., 3:4]

    return torch.cat([x_center, y_center, w, h], dim = -1)
```
